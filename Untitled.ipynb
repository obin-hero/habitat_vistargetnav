```python
import habitat

import cv2
import numpy as np
import matplotlib.pyplot as plt

from habitat.tasks.nav.shortest_path_follower import ShortestPathFollower
#from habitat.datas
import quaternion as q
```

```python
class VisualTargetNavEnv(habitat.RLEnv):
    def __init__(self, config):
        VisualTargetNavEnv.__init__(self, config)
        self.cube_size = 0.2
        self.target_radius = 0.5
        self.cube_resolution = 40
        self.cube_image = self.load_texture('wood.jpg', self.cube_resolution)
        
    def load_texture(self, texture_path, size):
        wood_texture = plt.imread(texture_path)
        self.texture_image = cv2.resize(wood_texture[:,42:208], (size,size))
        self.texture_points = np.array([[0.,0., size,size],[size,0.,0.,size]]).T
        
    def get_reward_range(self):
        return [-1, 1]

    def get_reward(self):
        return 0

    def get_done(self, observations):
        return self.habitat_env.episode_over

    def get_info(self, observations):
        return self.habitat_env.get_metrics()
    
```

```python
cfg = habitat.get_config('vistargetnav_mp3d.yaml')
cfg.defrost()
cfg.DATASET.DATA_PATH = '../habitat-api/data/datasets/pointnav/mp3d/v1/{split}/{split}.json.gz'
cfg.DATASET.SCENES_DIR = '../habitat-api/data/scene_datasets'
cfg.freeze()
print(cfg)
```

```python
env = VisualTargetNavEnv(cfg)
```
